```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
missing_values = ["n/a", "na", "--",".","NaN"]
dataset = pd.read_csv("openrefine/TrainingAll.csv", na_values = missing_values)
##dataset.fillna(-1)
```

```python
#list of collums
collist = dataset.columns.tolist()
```

```python
from sklearn import preprocessing
from sklearn.impute import SimpleImputer

robust_scaler = preprocessing.RobustScaler()
np_scaled = robust_scaler.fit_transform(dataset[collist[1:]])

fill_Nan = SimpleImputer(missing_values = np.nan, strategy ="constant", fill_value = -50.0)
np_imp = fill_Nan.fit_transform(np_scaled)
dataset_scaled = pd.DataFrame(np_imp)

# create the Labelencoder object
le = preprocessing.LabelEncoder()
dataset["File"] = le.fit_transform(dataset["File"])
dataset["File"].unique()

#merge the two dataframes
dataset_scaled["33"] = dataset["File"]
dataset_scaled.head(4)

```

```python
target_cols = ["CHGENROL","FRACDEGRE","RETPCNT"]
col_index = [dataset.columns.get_loc(name) for name in target_cols]
for index,col in enumerate(col_index):
    col_index[index] = col - 1
col_index
```

```python
#drop the seperate the data into targets and data

cols = [col for col in dataset_scaled.columns if col not in col_index]
X = dataset_scaled[cols]
y = dataset_scaled[col_index]
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
```

```python
def rfr_model(X, y):
# Perform Grid-Search
    gsc = GridSearchCV(
        estimator=RandomForestRegressor(),
        param_grid={
            'max_depth': range(10,50,5),
            'n_estimators': (10, 50, 100, 1000,2000,3000),
        },
        cv=5, scoring='neg_mean_squared_error', verbose=0, n_jobs=-1)
    
    grid_result = gsc.fit(X, y)
    best_params = grid_result.best_params_
    
    rfr = RandomForestRegressor(max_depth=best_params["max_depth"], n_estimators=best_params["n_estimators"],                               random_state=False, verbose=False)
# Perform K-Fold CV
    ##scores = cross_val_score(rfr, X, y, cv=10, scoring='neg_mean_absolute_error')
    ##predictions = cross_val_predict(rfr, X, y, cv=10)
    #return (scores,predictions,rfr)
    return rfr, best_params
```
